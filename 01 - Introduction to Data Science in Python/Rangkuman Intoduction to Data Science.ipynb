# Line Plot

```python
import numpy as np

x = np.array([-4,-3,-2,-1,0,1,2,3,4])

plt.style.use("fivethirtyeight")

plt.plot(x, x, label="y = x", color="c", linewidth=1, linestyle="-", marker="o")
plt.plot(x, x**2, label="y = x**2", color="m", linewidth=2, linestyle=":", marker="s")
plt.plot(x, x**3, label="y = x**3", color="y", linewidth=3, linestyle="--", marker="h")

plt.text(4.2, -2,"X")
plt.text(-4.1, 70,"Y")

plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.title("Grafik Fungsi", fontsize=20)

plt.legend()
plt.show()
```

# Scatter Plot

```python
cellphone = pd.read_csv("datasets/cell_phone_records.csv")

plt.scatter(cellphone.x, cellphone.y, alpha=0.1)

plt.xlabel("Langitude")
plt.ylabel("Latitude")

plt.show()
```

# Bar Plot

```python
x = [2011, 2012, 2013, 2014, 2015]
y = [499, 514, 621, 710, 654]

plt.bar(x,y)

plt.title("Students over 5 years")
plt.xlabel("Years")
plt.ylabel("Number of students")

plt.show()
```

```python
plt.barh(x,y)

plt.title("Students over 11 years")
plt.ylabel("Years")
plt.xlabel("Number of students")

plt.show()
```

```python
x = [2021,2022]
y1 = [4,5]
y2 = [6,8]

plt.bar(x, y1, label="dog")
plt.bar(x, y2, bottom=y1, label="cat")

plt.title("Number of dog and cats")
plt.xticks(x, ["2021", "2022"])

plt.legend()
plt.show()
```

# Histogram Plot

```python
gravius = pd.read_csv("datasets/shoe_gravel_sample.csv")

plt.hist(gravius.radius, bins=40, range=(2,8), density=True)

plt.xlabel('Gravel Radius (mm)')
plt.ylabel('Frequency')
plt.title('Sample from Shoeprint')

plt.show()
```
