# Random Float

```python
import numpy as np

np.random.seed(123) # pseudo-random numbers

angka1 = np.random.rand() # generate number between 0 and 1
print(angka1)
angka2 = np.random.rand()
print(angka2)
```

# Random Integer

```python
coin = np.random.randint(0,2) # generate int number between 0 until 1
print(coin)
if coin == 0:
    print("heads")
else:
    print("tails")
```

# Storing Data

```python
data = []
for i in range(10):
    coin = np.random.randint(0,2)
    if coin == 0:
        data.append("heads")
    else:
        data.append("tails")
print(data)
```

# Random Walk

### Example 1

```python
np.random.seed(123)

data = [0]
for i in range(10):
    coin = np.random.randint(0,2)
    data.append(data[-1] + coin)
print(data)
```

### Example 2

```python
# percobaan 10000 kali dengan setiap percobaan adalah 10 langkah
import matplotlib.pyplot as plt

np.random.seed(123)

data = []
for i in range(10000):
    tingkatan = [0]
    for j in range(10):
        coin = np.random.randint(0,2)
        tingkatan.append(tingkatan[-1] + coin)
    data.append(tingkatan[-1])

plt.hist(data,bins=10)
plt.show()
```

### Example 3

```python
# Empire State Building

# jika dadu bernilai 1 atau 2, maka step = step - 1
# jika dadu bernilai 3, 4, atau 5, maka step = step + 1
# jika dadu bernilai 6, maka step = step + np.random.randint(1,7)
# ada kemungkinan 0.1% jatuh ke step = 0. Dengan kata lain jika np.random.rand(0) <= 0, maka step = 0

data = []
for i in range(100):
    lantai = [0]
    for j in range(100):
        step = lantai[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0,step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        
        if np.random.rand() <= 0.001:
            step = 0
            
        lantai.append(step)
    data.append(lantai)

plt.plot(data) # data -> 2D np array -> x adalah baris dan y adalah kolom
plt.show()
plt.clf() # membuat lebih dari satu buah plot

data_t = np.transpose(data) # menukar elemen baris menjadi elemen kolom
plt.plot(data_t)
plt.show()
```
